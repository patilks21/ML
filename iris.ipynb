```python

df=pd.read_csv("iris.csv")
X = df.iloc[:, [0,1,2,3]].values
Y = df.iloc[:, 4].values
```

```python
data_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
data_rescaled = data_scaler.fit_transform(X)
data_rescaled
```

```python
binarizer = Binarizer(threshold=0.5).fit(X)
Data_binarized = binarizer.transform(X)
print("\nBinarty data:\n", Data_binarized[0:5])
```

```python
#standardscaler to convert scaled data with mean 0, SD =1
data = np.array([ [0,0],[1,0],[0,1],[1,1] ])
data_scaler = StandardScaler().fit(data)
data_rescaled = data_scaler.transform(data)
print(data_rescaled)
```
